```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
x = data['YearsExperience'].values.reshape(-1 , 1)
y = data['Salary']
```

```python
mod = LinearRegression()
```

```python
mod.fit(x,y)
```

```python
YearsExperience = float(input("Enter the years of Experience: "))
predict = mod.predict([[YearsExperience]])
```

```python
print("Your Salary is: ", predict[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(x,y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

plt.title('Exerience - Salary')

plt.show()

plt.plot(x,y)

sns.scatterplot(x='YearsExperience', y='Salary', data=data)
plt.plot(data['YearsExperience'], mod.predict(data[['YearsExperience']]), color="red")
```
