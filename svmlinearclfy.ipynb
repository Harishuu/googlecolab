```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
```

```python
X = np.array([1, 5, 2, 8, 1, 9, 7, 6.6, 2.3, 5.5, 7.7, 6.1])
y = np.array([2, 8, 1.8, 0.6, 11, 10, 9.4, 4, 3, 8.8, 7.5, 1])
```

```python
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
train_X = np.vstack((X, y)).T
train_y = [0, 1, 0, 1, 0, 1, 1, 1, 0, 0, 1, 1]
```

```python
clf = svm.SVC(kernel='linear', C=1.0)
```

```python
clf.fit(train_X, train_y)
```

```python
w = clf.coef_[0]  # weight calculation
a = -w[0] / w[1]  # Slope of the hyperplane
xaxis = np.linspace(0, 13)  # X-axis values for plotting
yaxis = a * xaxis - clf.intercept_[0] / w[1]  # Corresponding Y-axis values
```

```python
plt.plot(xaxis, yaxis, 'k-', label='Hyperplane')
```

```python
plt.scatter(train_X[:, 0], train_X[:, 1], c=train_y,)
plt.plot(xaxis,yaxis, 'k-')
plt.legend()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Linear SVM Classifier')
plt.show()
```
